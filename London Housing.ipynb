#### 1.1. Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 1.2.  Loading the data
data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.4.Transforming the data**

```python
clean_properties = pd.melt(properties_T, id_vars= ['London_Borough', 'ID'])
clean_properties.head()
```

```python
clean_properties = clean_properties.rename(columns = {0: 'Month', 'value': 'Average_price'})
clean_properties.head()
```

```python
clean_properties['Average_price'] = pd.to_numeric(clean_properties['Average_price'])
clean_properties.dtypes
```

```python
NaNFreeDF2 = clean_properties.dropna() 
NaNFreeDF2.head(48)
```

```python
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']
```

```python
NaNFreeDF2 = NaNFreeDF2[~NaNFreeDF2.London_Borough.isin(nonBoroughs)]
NaNFreeDF2.head()
```

```python
df = NaNFreeDF2
```

```python
df.head()
```

```python
df.dtypes
```

**2.6. Visualizing the data**

```python
df['Year'] = df['Month'].apply(lambda t: t.year)
```

```python
dfg = df.groupby(by=['London_Borough', 'Year']).mean(numeric_only=True)
dfg.sample(10)
```

```python
dfg = dfg.reset_index()
dfg.head()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2024 to the price in 2004.

```python
def create_price_ratio(d):
    y2003 = float(d['Average_price'][d['Year']==2003])
    y2023 = float(d['Average_price'][d['Year']==2023])
    ratio = [ y2003/y2023]
    return ratio
```

```python
create_price_ratio(dfg[dfg['London_Borough']=='Barking & Dagenham'])
```

```python
final={}
for b in dfg['London_Borough'].unique():
    borough = dfg[dfg['London_Borough'] == b]
    final[b] = create_price_ratio(borough)
print(final) 
```

```python
df_ratios = pd.DataFrame(final)
df_ratios.head()
```

```python
df_ratios = df_ratios.T
df_ratios = df_ratios.reset_index()
df_ratios.head()
```

```python
df_ratios.rename(columns={'index':'Borough', 0:'2023'}, inplace=True)
df_ratios.head()
```

```python
top15 = df_ratios.sort_values(by='2023',ascending=False).head(15)
print(top15)
```

```python
ax = top15[['Borough','2023']].plot(kind='bar')

ax.set_xticklabels(top15.Borough)
```
